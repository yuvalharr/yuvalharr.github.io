## Import packages

```r
library(readr)
library(psych)
library(tidyr)
library(ggplot2)
library(GGally)
library(lmerTest)
```

## Helper functions

```r
#sem: calculate standard error from the mean
sem <- function(x) sd(x)/sqrt(length(x))

#conversion between radian and degree
rad2deg <- function(rad) {(rad * 180) / (pi)}
deg2rad <- function(deg) {(deg * pi) / (180)}
```

### Figure 7a:
#### Visual crowding in two eccentricity (4 vs. 6) conditions

```r
visual_crowding = read_csv('../data/7ab_visual_crowding.csv')
```

```r
#Mean
aggregate(critical_spacing_deg~eccentricity_deg, data = visual_crowding, mean)

#SD
aggregate(critical_spacing_deg~eccentricity_deg, data = visual_crowding, sd)

#Standard error of the mean (sem)
aggregate(critical_spacing_deg~eccentricity_deg, data = visual_crowding, sem)

#wilcoxon rank sum test
wilcox.test(critical_spacing_deg~eccentricity_deg,data=visual_crowding)
```

### Figure 7b:
#### experiment accuracy

```r
visual_crowding_accuracy <- aggregate(correctRate~uuid, data = visual_crowding, mean)

#histgram of participant accuracy
ggplot(visual_crowding_accuracy, aes(x=correctRate)) + geom_histogram(aes(y=..density..), binwidth = 0.02, alpha=.8) +
  geom_vline(aes(xintercept=mean(correctRate)), color="red", size=1) +
  geom_density(alpha=.2, fill="grey") +
  scale_x_continuous(breaks = scales::pretty_breaks(n = 5)) +
  labs(title="", x="Percentage Correctness", y="Density") +
  theme_grey(base_size = 15)

#summary: mean, min, max
summary(visual_crowding_accuracy$correctRate)
```

### Table 2 & Figure 7a continued:
#### Mixed-effect model

```r
visual_crowding_dyslexia <- read_csv('../data/tab2_visual_crowding_dyslexia.csv')
```

```r
#stats for Fig. 7a
aggregate(critical_spacing_deg~eccentricity_deg+dyslexia, data = visual_crowding_dyslexia, mean)
```

```r
# quadratic model for Table 2
# dyslexics_cr_analysis <- merge(dyslexics_cr_analysis, new_demographics_0623, by=c('uuid','dyslexia'), all.x=TRUE)
visual_crowding_dyslexia$age_squared <- visual_crowding_dyslexia$age*visual_crowding_dyslexia$age
quadratic.model <- lmer(critical_spacing_deg~eccentricity_deg+dyslexia+age_squared+age+gender+(1|uuid),data=visual_crowding_dyslexia)
summary(quadratic.model)
```

```r
#Draw Fig. 7a
cr_bars <- aggregate(critical_spacing_deg~eccentricity_deg+dyslexia, data = visual_crowding_dyslexia, mean)
cr_stderr <- aggregate(critical_spacing_deg~dyslexia+eccentricity_deg, data = visual_crowding_dyslexia, sem)
names(cr_stderr)[3]<-paste("stderr")
cr_bars <- merge(cr_bars, cr_stderr, by=c('dyslexia','eccentricity_deg'))

ggplot(data=cr_bars, aes(x=as.factor(eccentricity_deg), y=critical_spacing_deg, fill=as.factor(dyslexia))) +
  geom_bar(stat="identity", position=position_dodge(), width = 0.5) + 
  geom_errorbar(aes(ymin=critical_spacing_deg-stderr, ymax=critical_spacing_deg+stderr), width=.1, position=position_dodge(.5)) +
  labs(x = "Eccentricity (deg)", y = "Visual Crowding (deg)") +
  scale_fill_manual("", labels = c("Non-dyslexic", "Dyslexic"), values=c("blue2","orange")) +
  theme(axis.text=element_text(size=17, color='black'), axis.title=element_text(size=19), legend.text=element_text(size=15)) +
  theme(legend.position = c(0.45, 0.9))
```

```r
#Mann-Whitney test between groups 
wilcox.test(critical_spacing_deg~eccentricity_deg,data=visual_crowding_dyslexia)
wilcox.test(critical_spacing_deg~dyslexia,data=subset(visual_crowding_dyslexia, eccentricity_deg == 6))
wilcox.test(critical_spacing_deg~dyslexia,data=subset(visual_crowding_dyslexia, eccentricity_deg == 4))
```

### Figure 7c:

### Figure 7d:
#### Within-subject standard deviation across three blind spot tests: mean, min, max of SD of calculated viewing distances. 

```r
viewing_distance_sd <- aggregate(viewDistance~uuid, data = viewing_distance, sd)

mean(viewing_distance_sd$viewDistance) #mean
min(viewing_distance_sd$viewDistance) #min
max(viewing_distance_sd$viewDistance) #max

#draw Fig. 7d
ggplot(viewing_distance_sd, aes(x=viewDistance)) + geom_histogram(aes(y=..density..), binwidth = 0.5, alpha=.8) +
  geom_vline(aes(xintercept=mean(viewDistance)), color="red", size=1) +
  geom_density(alpha=.2, fill="grey") +
  labs(title="", x="SD of Calculated Viewing Distance (cm)", y="Density")+
  theme_grey(base_size = 15)
```

### Figure 7e:
#### Within-subject viewing distance ICC and pairwise correlation.

```r
#import the df cleaned from `viewing_distance`
viewing_distance_by_session = read_csv('../data/7e_viewing_distance_by_session.csv')
```

```r
#transfer the df from long to wide
viewing_distance_wide <- spread(viewing_distance_by_session, session, viewDistance)
names(viewing_distance_wide)[2:4] <- c('s1','s2','s3')

#calculate ICC
ICC(viewing_distance_wide[,c(2,3,4)],missing=TRUE,alpha=.05)

#draw Fig. 7e
ggpairs(viewing_distance_wide, columns = 2:4, upper = list(continuous = wrap("cor", size=7, color='black'))) +
  theme(axis.text = element_text(size = 10))
```
