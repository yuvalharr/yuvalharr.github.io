### Import Packages

```r
library(readr)
library(psych)
library(tidyr)
library(ggplot2)
library(GGally)
library(lmerTest)
```

### Load data

18 participants, visual crowding (degrees) data from lab studies and online studies.

```r
exp_3_data <- read_csv("../data/exp_3_data.csv")
```

#### Draw Figure 6

```r

ggplot(exp_3_data, aes(x=lab_average, y=wild_average, color=as.factor(eccentricity))) + geom_point(size=3) + #geom_text(aes(label=subid),hjust=0, vjust=0)+
  geom_smooth(method=lm, se=FALSE, aes(group=1), fullrange=TRUE) +
  xlim(1,2.5) + ylim(1,2.5) +
  xlab("Visual Crowding in the Lab (deg)") + ylab("Visual Crowding in the Wild (deg)") +
  scale_color_discrete(name = "Eccentricity (deg)", labels = c("4", "6")) +
  theme_grey(base_size=15)
```

#### t-tests

```r
exp_3_data$avg = rowMeans(exp_3_data[c('wild_average', 'lab_average')])

# visual crowding in lab vs. in the wild --- no difference.
t.test(exp_3_data$wild_average, exp_3_data$lab_average, paired = TRUE)

# eccentricity 4 vs. 6
t.test(avg~eccentricity,data=exp_3_data)
```
